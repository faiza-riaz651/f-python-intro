```python
#car game
status=''
while status!='quit':
  cmd=input("""
  enter a command to start a self driving car..
  enter start to start
  enter stop to stop
  enter quit to exit
  COMMAND:
  """)
  if cmd.lower()=='stop':
    if status=='stop':
      print('already stopped')
    else:
      print('stop')
      status='stop'
  elif cmd.lower()=='start':
    if status=='stop':
      print('already started')
    else:
      print('start')
      status='start'
  elif cmd.lower()=='quit':
    # status='quit'
    break



  else:
    print('not a input i expected!!!!')
```

```python
#tuple,unpacking,dictionary, dictionary value access: by squre bracket and get

# program to convert digits to number in word
digit={1:'one',2:'two',3:'three',4:'four',5:'five',6:'six',7:'seven',8:'eight',9:'nine',0:'zero'}
x=input('enter the number:')
listt=[]
for num in range(len(x)):

  listt.append(digit.get(int(x[num])))
print(listt)
```
